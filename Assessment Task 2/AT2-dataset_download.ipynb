```python
import boto3
import sagemaker
import zipfile
from botocore.exceptions import ClientError

# ENTER YOUR STUDENT_ID BELOW --------------
student_Id = "25349642"
# ------------------------------------------

# ------- DON'T CHANGE THE CODE BELOW --------------
student_key = "student_"+student_Id +".zip"
key = "student-datasets-AT2/" + student_key
role = sagemaker.get_execution_role()
session = sagemaker.Session()
s3 = boto3.client('s3')
bucket_name = "sagemaker-student-datasets-443142193439-ap-southeast-2-an"
local_path = "./"+ student_key

try:
    print("✅ Starting Dataset Download...")
    s3.download_file(bucket_name, key, local_path)
    print("✅ Dataset " + student_key + " Download Complete!")
    
    # Unzip the dataset
    with zipfile.ZipFile(student_key, 'r') as zip:
        zip.extractall(".")
    print("✅ Unzip Done!")

except ClientError as e:
    if e.response['Error']['Code'] in ['404', 'NoSuchKey']:
        print(f"❌ Dataset for Student ID '{student_Id}' was not found.")
        print("   Please check your Student ID is correct and try again.")
        print("   If the problem persists, contact the Subject Coordinator.")
    elif e.response['Error']['Code'] == '403':
        print(f"❌ Access denied for Student ID '{student_Id}'.")
        print("   Please contact the Subject Coordinator.")
    else:
        print(f"❌ Unexpected error: {e}")
        print("   Please contact the Subject Coordinator.")
```
